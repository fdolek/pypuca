```python
def func(t, a, b):
    return a * (np.exp((4-t)*b)) 
```

```python
ydata = np.array([1,1,1,5,6,18,47,98,191,359,670,947,1236,1529,1872,2433,3629,5698,7402,9217,10827,13531,15679,18135,20921,23934,27069,30217,34109,38226,42282,47029,52167,56956,61049,65111])
xdata = np.arange(len(ydata))
```

```python
plt.plot(ydata, 'b-', label='Turkey',linewidth=3)
popt, pcov = curve_fit(func, xdata, ydata, p0=(200.,-0.1))

plt.plot(xdata,func(xdata, *popt),
                             'r-.',
                             label=r'$fit: a=%5.3f, \tau=%5.3f, $' % tuple(popt),
                             linewidth=2)
plt.ylim(-1000, 100000)
plt.xlim(0, 50)
plt.legend()
```

```python
import matplotlib.pyplot as plt
from scipy import interpolate
y =  np.array([1,1,1,5,6,18,47,98,191,359,670,947,1236,1529,1872,2433,3629,5698,7402,9217,10827,13531,15679,18135,20921,23934,27069,30217,34109,38226,42282])
x = np.arange(len(y))
f = interpolate.interp1d(x, y)
xnew = np.arange(0, 30, 0.1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.ylim(-1000, 100000)
plt.xlim(0, 50)
plt.show()
```
